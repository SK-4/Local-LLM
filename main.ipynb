```python
import nest_asyncio
from gitingest import ingest

# Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()

# Function to process the GitHub repository using GitIngest (synchronously)
def process_with_gitingest(github_url):
    summary, tree, content = ingest(github_url)
    return summary, tree, content

# Example usage for both 

# Example GitHub URL
github_url = "https://github.com/SK-4/Calmly"

# Step 2: Process GitHub URL with GitIngest
summary, tree, content = process_with_gitingest(github_url)
print(f"GitHub Summary: {summary}")
print(f"GitHub Tree: {tree}")
print(f"GitHub Content: {content}")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM

template = """Question: {question}

Answer: Let's think step by step."""

prompt = ChatPromptTemplate.from_template(template)

model = OllamaLLM(model="llama3.2")

chain = prompt | model

chain.invoke({"question": "What is LangChain?"})
```
